```python
sorted_by_Deaths = covid_df.sort_values(by='Deaths' ,ascending=False)
sorted_by_Deaths.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(sorted_by_Deaths['Country/Region'],sorted_by_Deaths['Deaths'])
```

```python
top10_Deaths=sorted_by_Deaths.iloc[:10,0:10]
top10_Deaths
```

```python
plt.figure(figsize=(10,5))
plt.bar(top10_Deaths['Country/Region'],top10_Deaths['Deaths'])
```

```python
sorted_by_NewCases = covid_df.sort_values(by='New cases' ,ascending=False)
sorted_by_NewCases.head()

```

```python
plt.figure(figsize=(10,5))
plt.bar(top10_NewCases['Country/Region'],top10_NewCases['New cases'])
```

```python
sorted_by_Recovered = covid_df.sort_values(by='Recovered' ,ascending=False)
sorted_by_Recovered.head()
```

```python
sorted_by_Confirmed = covid_df.sort_values(by='Confirmed' ,ascending=False)
sorted_by_Confirmed.head()
```

```python
top10_Confirmed=sorted_by_Confirmed.iloc[:10,0:10]
plt.figure(figsize=(10,5))
plt.bar(top10_Confirmed['Country/Region'],top10_Confirmed['Confirmed'])
```

```python
top10_cofirmed_by_Deaths = top10_Confirmed.sort_values(by='Deaths' ,ascending=False)
plt.figure(figsize=(10,5))
plt.bar(top10_cofirmed_by_Deaths['Country/Region'],top10_cofirmed_by_Deaths['Deaths'])
```

```python
top10_cofirmed_by_Recovered = top10_Confirmed.sort_values(by='Recovered' ,ascending=True)
top10_cofirmed_by_Recovered.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(top10_cofirmed_by_Recovered['Country/Region'],top10_cofirmed_by_Recovered['Recovered'])
```

```python
top10_cofirmed_by_Newcases = top10_Confirmed.sort_values(by='New cases' ,ascending=False)
plt.figure(figsize=(10,5))
plt.bar(top10_cofirmed_by_Newcases['Country/Region'],top10_cofirmed_by_Newcases['New cases'])
```
